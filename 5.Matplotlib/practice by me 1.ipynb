```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x=np.linspace(0,10,15)
```

```python
y=x**2
```

```python
#functional
plt.plot(x,y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Title')
```

```python
#multiplot
plt.subplot(1,2,1)
plt.plot(x,y,'r-')

plt.subplot(1,2,2)
plt.plot(y,x,'b-')
```

```python
#object oriented method
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X axis')
axes.set_ylabel('Y axis')
axes.set_title('Title')
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y,'r')
axes2.plot(y,x)
axes1.set_title('Large plot')
axes2.set_title('Smaller plot')
```

```python
#OOP subplots
fig,axes=plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[0].set_title('First Title')
axes[1].plot(y,x)
axes[1].set_title('Second Title')

plt.tight_layout()
```

```python
#figure size and Dpi
fig=plt.figure(figsize=(5,2))
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes=plt.subplots(nrows=2,ncols=1,figsize=(5,3))
axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

```python
#Save figure
fig.savefig('my_picture.png',dpi=200)
```

```python
#legend
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(x,y,label='Y normal')
axes.plot(x,y*2,label='Y squre')

axes.legend(loc=0)
```
